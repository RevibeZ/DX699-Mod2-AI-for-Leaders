Thus:

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe()
```

8. What are the mean, median, and mode customer age in years?  (Rounding down to the next lower age.)
Are there any outliers?  (Customers with very large or very small ages, compared with the other ages?)

Suggested Google search or ChatGPT prompt: "how can I find out the mean, median, and mode of a pandas Series"

```python
# Using the most "cleaned" dataframe available
df_age_base = df_accountid_unique.copy()

# Choose a reference date for "age" (Today's date)
ref_date = pd.Timestamp.today().normalize()

birth_dates = df_age_base["BirthDate"]

age_years = ((ref_date - birth_dates).dt.days // 365).astype("Int64")
age_years_clean = age_years.dropna().astype(int)

print("Reference date:", ref_date.date())
print("N ages (non-missing):", len(age_years_clean))

# Statistical Data calculations/print outs
mean_age = age_years_clean.mean()
median_age = age_years_clean.median()
mode_ages = age_years_clean.mode().tolist()

print("Mean age (years):", mean_age)
print("Median age (years):", median_age)
print("Mode age(s) (years):", mode_ages)

# Outliers using the IQR rule
q1 = age_years_clean.quantile(0.25)
q3 = age_years_clean.quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outlier_mask = (age_years_clean < lower) | (age_years_clean > upper)
outlier_ages = age_years_clean[outlier_mask]

print(f"\nIQR outlier bounds: [{lower:.1f}, {upper:.1f}]")
print("Number of outlier ages:", len(outlier_ages))

if len(outlier_ages) > 0:
    outlier_rows = df_age_base.loc[outlier_ages.index, ["CustomerID", "BirthDate", "AccountID", "AccountType"]].copy()
    outlier_rows["AgeYears"] = outlier_ages.values
    outlier_rows.sort_values("AgeYears").head(10)
else:
    print("No outliers detected by the IQR rule.")
```

```python
# Problem 10: sanity checks for "values that don't seem right"
df_check = df2.copy()

# Reference Date for age calculations
ref_date = pd.Timestamp.today().normalize()

# 1) BirthDate older than 120 years
age_years = ((ref_date - df_check["BirthDate"]).dt.days // 365).astype("Int64")
too_old = age_years > 120
print("BirthDate > 120 years ago:", int(too_old.sum()))
if too_old.any():
    display(df_check.loc[too_old, ["CustomerID", "BirthDate", "AccountID", "AccountOpened", "AccountType"]].head())

# 2) BirthDate too recent
very_young = age_years < 4
print("Customers younger than 4:", int(very_young.sum()))
if very_young.any():
    display(df_check.loc[very_young, ["CustomerID", "BirthDate", "AccountID", "AccountOpened", "AccountType"]].head())

# 3) AccountOpened in the future relative to ref_date
opened_in_future = df_check["AccountOpened"] > ref_date
print("Accounts opened after reference date:", int(opened_in_future.sum()))
if opened_in_future.any():
    display(df_check.loc[opened_in_future, ["CustomerID", "BirthDate", "AccountOpened", "AccountID", "AccountType"]].head())

# 4) AccountOpened before BirthDate
opened_before_birth = df_check["AccountOpened"] < df_check["BirthDate"]
print("Accounts opened before birth:", int(opened_before_birth.sum()))
if opened_before_birth.any():
    display(df_check.loc[opened_before_birth, ["CustomerID", "BirthDate", "AccountOpened", "AccountID", "AccountType"]].head())

# 5) "Two-year-olds opening accounts"
age_at_open_years = ((df_check["AccountOpened"] - df_check["BirthDate"]).dt.days // 365).astype("Int64")
toddlers_opening = age_at_open_years < 10
print("Accounts opened when customer < 10 years old:", int(toddlers_opening.sum()))
if toddlers_opening.any():
    display(df_check.loc[toddlers_opening, ["CustomerID", "BirthDate", "AccountOpened", "AccountID", "AccountType"]].head())
```

Here is an example:

```python
import matplotlib.pyplot as plt

ax = plt.gca() # get an "Axes" object to draw on; gca stands for "get current Axes"
ax.scatter(df2["BirthDate"], df2["AccountOpened"]) # create a scatter plot based on these two dates
ax.set_ylabel("Account Opened") # label the y axis
ax.set_xlabel("Birth Date") # label the x axis
```

```python
# Q11: Analyze age at account opening vs. birth year
df_q11 = df2.copy()

# Compute age-at-open (rounded down to whole years)
df_q11["AgeAtOpenYears"] = ((df_q11["AccountOpened"] - df_q11["BirthDate"]).dt.days // 365).astype("Int64")
df_q11["BirthYear"] = df_q11["BirthDate"].dt.year.astype("Int64")

# Keep rows with valid ages and a reasonable range for plotting
df_plot = df_q11.dropna(subset=["BirthYear", "AgeAtOpenYears"]).copy()
df_plot = df_plot[(df_plot["AgeAtOpenYears"] >= 0) & (df_plot["AgeAtOpenYears"] <= 120)]

print("Rows available for plotting:", len(df_plot))
print("Age-at-open summary (years):")
display(df_plot["AgeAtOpenYears"].describe())

# Simple numeric association
corr = df_plot["BirthYear"].astype(int).corr(df_plot["AgeAtOpenYears"].astype(int))
print("Correlation(BirthYear, AgeAtOpenYears):", round(float(corr), 3))

# Build a binned mean line (mean age-at-open by birth-year bins)
bins = np.arange(int(df_plot["BirthYear"].min()), int(df_plot["BirthYear"].max()) + 6, 5)  # 5-year bins
df_plot["BirthYearBin"] = pd.cut(df_plot["BirthYear"].astype(int), bins=bins, right=False)
bin_means = df_plot.groupby("BirthYearBin", observed=True)["AgeAtOpenYears"].mean()
bin_centers = [interval.left + 2.5 for interval in bin_means.index]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Birth year vs age at open
axes[0].scatter(df_plot["BirthYear"].astype(int), df_plot["AgeAtOpenYears"].astype(int), alpha=0.5, s=25)
axes[0].plot(bin_centers, bin_means.values, color="red", linewidth=2, label="Mean age-at-open (5-year bins)")
axes[0].set_title("Age at Account Opening vs. Birth Year")
axes[0].set_xlabel("Birth year")
axes[0].set_ylabel("Age when account opened (years, floored)")
axes[0].grid(True, alpha=0.25)
axes[0].legend()

# Plot 2: Histogram of age at opening
axes[1].hist(df_plot["AgeAtOpenYears"].astype(int), bins=20, edgecolor="white")
axes[1].set_title("Distribution of Ages at Account Opening")
axes[1].set_xlabel("Age when account opened (years, floored)")
axes[1].set_ylabel("Number of customers")
axes[1].grid(True, axis="y", alpha=0.25)

plt.tight_layout()
plt.show()
```

### Q11 Interpretation 
- The scatter plot (Birth year vs. age at account opening) shows a wide spread of ages for nearly every birth-year range, with no clear upward/downward pattern.
- The correlation between `BirthYear` and `AgeAtOpenYears` is very close to 0 (about `0.038`), which suggests little to no linear relationship in this dataset.
- The histogram shows ages at account opening are broadly distributed (roughly 0–79 years here), not concentrated in a narrow “typical” opening age.

**Conclusion:** In this (synthetic) dataset, there does not appear to be a strong connection between the year someone is born and the age at which they open their account.

```python
# Data
labels = ["Excited", "Kind of interested", "OK", "Not Great", "Bored"]
sizes = [19, 25, 40, 5, 11]

# Pick colors for each category
color_by_label = {
    "Bored": "#388ae8",
    "Not Great": "#af3347",
    "OK": "#90dd49",
    "Kind of interested": "#604ce4",
    "Excited": "#5cd1cd",
}

colors_pie = [color_by_label[label] for label in labels]

# Draw Pie Chart
fig, ax = plt.subplots(figsize=(6, 4))
wedges_pie, texts, autotexts = ax.pie(
    sizes,
    colors=colors_pie,
    startangle=90,
    autopct="%1.0f%%",
    pctdistance=0.8,
 )

# Titles
ax.set_title("Survey Results")
ax.axis("equal") 

# Legend corresponding to each category
ax.legend(
    wedges_pie[::-1],
    labels[::-1],
    loc="center left",
    bbox_to_anchor=(1.0, 0.5),
 )

plt.tight_layout()
plt.show()
```
