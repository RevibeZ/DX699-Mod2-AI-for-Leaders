# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The side by side plots make it easier to focus on the shape and trend of each week individually. They reduce visual clutter and avoid overlap, so each week can be read clearly on its own. However, the disadvantage is that direct comparison is harder because the viewer has to shift their eyes back and forth between the two panels, which makes differences by day less immediate.

The combined line plot makes day to day comparison between the two weeks much easier because both series share the same axes. This helps the viewer quickly see where one week had more visitors, where the lines cross, and how the two patterns differ overall. The disadvantage is that the chart can become cluttered if more lines are added, and overlapping lines may make it harder to focus on the pattern of either week individually. 

Overall, the combined plot is better for direct comparison, while the separate plots are better for emphasizing each week as its own pattern.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
years = np.array([2019, 2020, 2021, 2022, 2023, 2024, 2025])

product_a = [395, 420, 423, 390, 300, 270, 260]
product_b = [360, 400, 410, 375, 290, 260, 250]
product_c = [np.nan, np.nan, 100, 180, 200, 240, 180]
product_d = [np.nan, np.nan, np.nan, 160, 260, 220, 215]
product_e = [np.nan, np.nan, np.nan, np.nan, np.nan, 95, 210]

plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.set_facecolor("#f5f5f5")
plt.gcf().patch.set_facecolor("#f5f5f5")

plt.plot(years, product_a, color="#4F81BD", linewidth=4, solid_capstyle="round", label="Product A")
plt.plot(years, product_b, color="#C0504D", linewidth=4, solid_capstyle="round", label="Product B")
plt.plot(years, product_c, color="#9BBB59", linewidth=4, solid_capstyle="round", label="Product C")
plt.plot(years, product_d, color="#8064A2", linewidth=4, solid_capstyle="round", label="Product D")
plt.plot(years, product_e, color="#4BACC6", linewidth=4, solid_capstyle="round", label="Product E")

plt.title("Average Retail Product Price Per Year", fontsize=24, weight="bold", pad=18)
plt.xticks(years, fontsize=12)
plt.yticks(np.arange(0, 501, 100), [f"${y}" for y in range(0, 501, 100)], fontsize=12)
plt.ylim(0, 500)

ax.grid(axis="y", color="#bfbfbf", linewidth=1)
ax.grid(axis="x", visible=False)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#9e9e9e")
ax.spines["bottom"].set_color("#9e9e9e")
ax.tick_params(color="#9e9e9e", labelcolor="black", length=5)

plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.13),
    ncol=5,
    frameon=False,
    fontsize=12,
    handlelength=1.8,
    handletextpad=0.3
)

plt.tight_layout()
plt.show()
```
