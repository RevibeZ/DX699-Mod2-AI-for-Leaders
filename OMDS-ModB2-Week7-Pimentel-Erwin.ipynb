# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

To draw the viewer’s attention to the last three months of data first, several preattentive attributes could be used to visually emphasize that portion of the graph. One approach would be to highlight the final three data points using a different color while keeping the earlier months in a neutral gray tone. This color contrast would immediately direct attention to the most recent values. Another option would be to increase the size of the markers for the final three months or add labels directly to those points to make them stand out. A shaded background region behind the final three months could also be used to separate them visually from the rest of the timeline. Finally, annotations or callout text could be placed near the last data points to guide the viewer toward the most recent trend. Each of these approaches uses preattentive visual cues to direct attention quickly and effectively.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Data
concerns = [
    "Engine power is less than expected",
    "Tires make excessive noise while driving",
    "Engine makes abnormal/excessive noise",
    "Seat material concerns",
    "Excessive wind noise",
    "Hesitation or delay when shifting",
    "Bluetooth system has poor sound quality",
    "Steering system/wheel has too much play",
    "Bluetooth system is difficult to use",
    "Front seat audio/entertainment/navigation controls"
]

values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]

# Reverse for barh so highest appears at top
concerns_plot = concerns[::-1]
values_plot = values[::-1]

colors_plot = [
    "#8f8f8f", "#8f8f8f", "#8f8f8f",   # bottom 3 in final chart
    "#c7524a", "#c7524a", "#c7524a", "#c7524a",
    "#c7524a", "#c7524a", "#c7524a"    # top 7 in final chart
]

plt.figure(figsize=(11, 6.2))
ax = plt.gca()

bars = ax.barh(concerns_plot, values_plot, color=colors_plot, height=0.76)

# Value labels inside bars
for bar, val in zip(bars, values_plot):
    ax.text(
        val - 0.25,
        bar.get_y() + bar.get_height() / 2,
        f"{val:.1f}",
        va="center",
        ha="right",
        color="white",
        fontsize=12,
        fontweight="bold"
    )

# Title and subtitle
ax.set_title("Top 10 design concerns", loc="left", fontsize=18, color="#666666", pad=18)
ax.text(0.56, 1.01, "concerns per 1,000",
        transform=ax.transAxes, fontsize=12, color="#777777", ha="center")

# Clean up axes
ax.set_xlim(0, 13.2)
ax.set_xticks([])
ax.set_xlabel("")
ax.set_ylabel("")

for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(axis="y", labelsize=12, colors="#555555")

plt.tight_layout()
plt.show()
```
